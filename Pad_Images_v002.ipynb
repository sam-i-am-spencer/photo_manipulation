**Notes:**
- We want to pad in a way the will make the most of our 4x6 printing area. Place the longest dimension of the original image along the 6" side and pad the shorter to 4".
- We don't want to reduce the original image

```python
def expand_to_4x6_photo(pil_img, background_color=(255,255,255)):
    '''
    Function to pad out meme's so they can be easily printed on 4x6" paper.
    '''
    # Get the width and height
    width, height = pil_img.size
    orig_area = width * height
    print('---Original Image Dimensions---')
    print('width: {} and height: {}\n'.format(width, height))

    # If the width and height ratio already 4x6 return the original image
    if width == (height/4)*6 or height == (width/4)*6:
        print('Do nothing')
        return pil_img
    # Examine the effect of increasing the width or height to get the correct 4x6 aspect ratio
    else:
        # Scenario 1 - Increase the height
        height_new = int(round((width/4)*6, 0))
        area_1 = width*height_new
        print('---Increased height---')
        print('width: {} and height: {}'.format(width, height_new))
        print('Area 1 diff: {}\n'.format(area_1-orig_area))
        
        # Scenario 2 - Increase the width
        width_new = int(round((height/4)*6, 0))
        area_2 = width_new*height
        print('---Increased width---')
        print('width: {} and height: {}'.format(width_new, height))
        print('Area 2 diff: {}\n'.format(area_2-orig_area))
        
        
        if (area_1-orig_area) < (area_2-orig_area) and height_new > height:
            # Increase the height dimension
            result = Image.new(pil_img.mode, (width, height_new), background_color)
            result.paste(pil_img, (0, ((height_new - height) // 2)))
            #result.paste(pil_img, (0, 0))
            print('Increasing the height')
            print('new dimensions are width: {} and height: {}'.format(width, height_new))
            return result
        elif (area_2-orig_area) < (area_1-orig_area) and width_new > width:
            # Increase the width dimension
            result = Image.new(pil_img.mode, (width_new, height), background_color)
            #result.paste(pil_img, (29, 0))
            result.paste(pil_img, (((width_new - width) // 2), 0))
            print('Increasing the width')
            print('new dimensions are width: {} and height: {}'.format(width_new, height))
            return result
        elif (area_2-orig_area) < 0:
            # Increase the height dimension
            result = Image.new(pil_img.mode, (width, height_new), background_color)
            result.paste(pil_img, (0, ((height_new - height) // 2)))
            #result.paste(pil_img, (0, 0))
            print('Increasing the height')
            print('new dimensions are width: {} and height: {}'.format(width, height_new))
            return result
        else:
            # Increase the width dimension
            result = Image.new(pil_img.mode, (width_new, height), background_color)
            #result.paste(pil_img, (29, 0))
            result.paste(pil_img, (((width_new - width) // 2), 0))
            print('Increasing the width')
            print('new dimensions are width: {} and height: {}'.format(width_new, height))
            return result
        
    print('--------------------------\n')
```

```python
# r=root, d=directories, f = files
for r, d, f in os.walk(abs_src_path):
    for file in f:
        # Look for .jpg|.JPG|.jpeg|.JPEG
        if any(x in file.lower() for x in [".jpg", ".JPG", ".jpeg", ".JPG"]):
            print('Processing Image: {}'.format(file))
            src = os.path.join(r, file)
            image_file = Image.open(src)
            # Pad the meme
            im_new = expand_to_4x6_photo(image_file)
            file_name = file[:-4] + '_4x6' + ".jpg"
            #im_new.save(abs_dst_path + '/' + file_name, quality=100)
            im_new.save(os.path.join(abs_dst_path, file_name), quality=100)
            print(abs_dst_path + '/' + file_name)
            print('\n')
```
